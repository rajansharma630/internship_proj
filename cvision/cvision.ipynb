```python
import cv2
import numpy as np
image=cv2.imread('lio2.jpg')
model = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
faces = model.detectMultiScale(image)
print(faces)
for face in faces:
    x,y,w,h=face
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3)
cv2.imwrite('le02ans.jpg',image)
```

```python
import cv2
import numpy as np

# Read image
image = cv2.imread("lio2.jpg")

# Load Haar Cascade
model = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

print(faces)

# Draw rectangles
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 3)

# Save output
cv2.imwrite("le02ans.jpg", image)

print("Face detection completed successfully.")
```
